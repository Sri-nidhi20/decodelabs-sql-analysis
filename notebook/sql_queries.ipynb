## Query 1: Total Number of Orders

### Business Question

How many orders are present in the dataset?

```python
query1 = """
SELECT Count(*) AS Total_Orders
From df
"""

result1 = pysqldf(query1)

result1
```

### Insight

The dataset contains 1200 orders, representing the total transaction volume available for analysis. This provides a sufficient sample size to identify customer purchasing patterns, revenue trend, and operational performance metrics.

## Query 2: Revenue Generated by Each Product

### Business Question

Which products generate the highest revenue?

```python
query2 = """
SELECT Product,
    Round(SUM(TotalPrice), 2) AS Revenue
FROM df
GROUP BY Product
ORDER BY Revenue DESC
"""

result2 = pysqldf(query2)

result2
```

### Insight

Chair generated the highest revenue (195,620.11/-), closely followed by Printer (195,612.61/-) and Laptop (192,126.56/-). These products contribute significantly to overall business revenue and should be prioritized for inventory management, promotions, and customer targeting strategies. Phone generated the lowest revenue among the listed products, indicating comparatively lower sales performance.

## Query 3: Average Order Value

### Business Question

What is the average value of an order?

```python
query3 = """
SELECT ROUND(AVG(TotalPrice), 2) AS Average_Order_Value
FROM df
"""

result3 = pysqldf(query3)

result3
```

### Insight

The average order value across all transactions is 1053.97/-. This indicates that customers typically spend around 1000/- per purchase. Increasing average order value through bundle offers, product recommendations, and upselling strategies could significantly improve overall revenue without increasing customer acquisition costs.


Since the average order value is above 1000/-, the business appears to be dealing with medium-to-high value transactions. Marketing campaigns focused on premium and cross-selling opportunities may generate higher return compared to discount-focused strategies.

## Query 4: Orders by Payment Method

### Business Question

Which payment methods are most frequently used by customers?

```python
query4 = """
SELECT
    PaymentMethod,
    COUNT(*) AS Total_Orders
FROM df
GROUP BY PaymentMethod
ORDER BY Total_Orders DESC
"""

result4 = pysqldf(query4)

result4
```

### Insight

Online payment is the most frequently used payment method  with 258 orders, followed by Cash (246 orders), Credit Card (234 orders), Debit Card (232 orders), and Gift Cards (230 orders).

The distribution is fairly balanced across all payment methods, indicating that customers prefer multiple payment options rather than relying on single mode of payment.


The company should continue supporting all major payment methods since customer usage is distributed across different options. However, optimizing the online payment experience could further improve customer convenience and checkout efficiency.

## Query 5: Order Status Analysis

### Business Question

What is the distribution of customer order statuses?

```python
query5 = """
SELECT 
    OrderStatus,
    COUNT(*) AS Total_Orders
FROM df
GROUP BY OrderStatus
ORDER BY Total_Orders DESC
"""

result5 = pysqldf(query5)

result5
```

### Insight

Order Statuses are distributed relatively evenly across all categories, including delivered, shipped, pending, returned, and cancelled orders.

This balanced distribution provides a useful dataset for analyzing different stages of the order lifecycle and understanding how order outcomes may impact overall business performance.


The company should continuously monitor order status trends and identify factors contributing to cancellations, returns and pending orders. Reducing unsuccessful order outcomes can improve customer satisfaction and operational efficiency.

## Query 6: Top Customers by Revenue

### Business Question

Which customers generate the highest revenue for the business?


```python
query6 = """
SELECT
    CustomerID,
    ROUND(SUM(TotalPrice),2) AS Revenue
FROM df
GROUP BY CustomerID
ORDER BY Revenue DESC
LIMIT 10
"""

result6 = pysqldf(query6)

result6
```

### Insight

Customer C38840 generated the highest revenue of 5723.23/-, significantly higehr than other customers. The remaining top customers contributed between 3313/- and 3456/- in revenue.

This indicates the presence of high-value customers who contribute substantially to overall business revenue.

The business should focus on retaining high-value customers through loyalty programs, exclusive offers, and personalized marketing campaigns. Customer retention is generally more cost-effective than acquiring new customers.

## Query 8: Products by Quantity Sold

### Business Question

Which products are sold most frequently?

```python
query8 = """
SELECT
    Product,
    SUM(Quantity) AS Total_Quantity_Sold
FROM df
GROUP BY Product
ORDER BY Total_Quantity_Sold DESC
"""

result8 = pysqldf(query8)

result8
```

### Insight

Chair recorded the highest quantity sold (562 units), followed by Printer (542 units) and Laptop (535 units). These products demomstrate strong customer demand and contribute significantly to sales volume.

Phone recorded the lowest quantity sold (411 units), indicating comparatively lower demand.


The company should prioritize inventory availability for high-demand products such as Chairs, Printers and Laptops to prevent stock shortages. Marketing campaigns and promotional offers can be used to improve sales of lower-performing products.

## Query 9: Revenue by Order Status

### Business Question

Which order statuses contribute the most revenue?

```python
query9 = """
SELECT 
    OrderStatus,
    ROUND(SUM(TotalPrice),2) AS Revenue
FROM df
GROUP BY OrderStatus
ORDER BY Revenue DESC
"""

result9 = pysqldf(query9)

result9
```

### Insight

Revenue is distributed relatively evenly across all order statuses. Cancelled orders generated the highest recorded revenue (276,396.21/-), followed by Pending (256,328.15/-), Shipped (246,159.58/-), Returned (243,277.70/-), and Delivered orders (242,600.32/-).

The balanced distributed indicates that the dataset contains diverse transaction scenarios, making it suitable for analytical and SQL practice.

Organizations should closely monitor order lifecycle stages and implement strategies to convert pending orders into completed deliveries while minimizing cancellations and returns. Improved order fulfillment processes can positively impact revenue generation and customer satisfaction.

## Query 10: Referral Source Performance

### Business Question

Which referral sources bring the highest number of customers?

```python
query10 = """
SELECT
    ReferralSource,
    COUNT(*) AS Total_Customers
FROM df
GROUP BY ReferralSource
ORDER BY Total_Customers DESC
"""

result10 = pysqldf(query10)

result10
```

### Insight

Instagram generated the highest number of customer acquisitions (259 customers), followed by Email (250) and Google (241).

This indicates that social media and digital marketing channels play a significant role in attracting customers and driving business growth.


The company should continue investing in high-performing acquisition channels such as instagram and Email marketing while optimizing lower-performing channels to improve customer reach and marketing efficiency.
